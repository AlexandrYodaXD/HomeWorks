```python
# задаю переменной x как математическую переменную
x = Symbol('x')

# задаю функцию f(x)
f = -12*x**4*sin(cos(x))-18*x**3+5*x**2+10*x-30
f
```

<h1><ins> Из графика функции можно сделать вывод что функция имеет бесконечное колличество корней.</ins></h1>

<h3> В таком случае, следующие критерии задания можно выполнить только на каком-либо конкретном участке функции:</h3>

- Определить корни

- Найти интервалы, на которых функция возрастает

- Найти интервалы, на которых функция убывает

- Вычислить вершину

- Определить промежутки, на котором f > 0

- Определить промежутки, на котором f < 0

<h3>Создам две вспомогательные функции.</h3>

```python
# функция f(x)
def func_1(x):
    return -12 * (x ** 4) * np.sin(np.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30

# функция f(x) = 0
def func_2(x, y):
    return -12 * (x ** 4) * np.sin(np.cos(x)) - 18 * x ** 3 + 5 * x ** 2 + 10 * x - 30 - y
```

<h3>Возьму для рассмотрения участок графика функции f(x) с таким диапазоном x, где будут выполняться все необходимые критерии задания.</h3>

```python
x_left, x_right = -1.4, 2.3
plot(f, (x, x_left, x_right))
```

<h3>Далее для отрисовки графика функции буду использовать matplotlib</h3>

```python
# добавлю оси на координатную плоскость
plt.axhline(0,color='black')
plt.axvline(0,color='black')

# график f(x) при x [-1.4, 2.3]
full = np.arange(x_left, x_right, 0.01)
plt.plot(full, func_1(full), 'b-.')
```

<h3>Из графика видно что на данном участке функция имеет два корня. Найдём их.</h3>

```python
# первый корень лежит в промежутке от -2 до 1
root_1 = fsolve(func_1, -2)
root_1 = (*root_1,)[0]
print('Первый корень:', root_1)

# второй корень лежит в промежутке от 2 до 3
root_2 = fsolve(func_1, 2)
root_2 = (*root_2,)[0]
print('Второй корень:', root_2)
```

<h3>Найдем вершины графика.</h3>

```python
# первая вершина лежит в промежутке x[-1, 0]
vertex_1_arr = np.arange(-1, 0, 0.001)
vertex_1_y = min(func_1(vertex_1_arr))
vertex_1_x = fsolve(func_2,-1, vertex_1_y)[0]
print(f'Первая вершина: ({vertex_1_x}, {vertex_1_y})')

# вторая вершина лежит в промежутке x[0, 1]
vertex_2_arr = np.arange(0, 1, 0.001)
vertex_2_y = max(func_1(vertex_2_arr))
vertex_2_x = fsolve(func_2, 0, vertex_2_y)[0]
print(f'Вторая вершина: ({vertex_2_x}, {vertex_2_y})')

# третяя вершина лежит в промежутке x[1, 2]
# она же самая нижняя точка на рассматриваемом участке графика
vertex_3_arr = np.arange(1, 2, 0.001)
vertex_3_y = min(func_1(vertex_3_arr))
vertex_3_x = fsolve(func_2, 1, vertex_3_y)[0]
print(f'Третья вершина / нижняя точка рассматриваемого участка графика: ({vertex_1_x}, {vertex_3_y})')
```

<h3>Покажу всё выше перечисленное на одном графике</h3>

```python
# добавлю оси на координатную плоскость
plt.axhline(0,color='black')
plt.axvline(0,color='black')

# корень 1
plt.plot(root_1, 0, 'go', label='Корни')

# корень 2
plt.plot(root_2, 0, 'go')

# убывает при y > 0
pos_down_1 = np.arange(x_left, root_1, 0.01)
plt.plot(pos_down_1, func_1(pos_down_1), 'r-', label='↓ при y > 0')

# первая вершина
plt.plot(vertex_1_x, vertex_1_y, 'rx', label='Вершины')

# убывает от первого корня до первой вершины
neg_down_1 = np.arange(root_1, vertex_1_x, 0.01)
plt.plot(neg_down_1, func_1(neg_down_1), 'y-', label='↓ от 1-го корня до 2-ой вершины')

# вторая вершина
plt.plot(vertex_2_x, vertex_2_y, 'rx')

# возрастает от первой вершины до второй вершины
neg_up_1 = np.arange(vertex_1_x, vertex_2_x, 0.01)
plt.plot(neg_up_1, func_1(neg_up_1), 'g-', label='↑ от 1-ой до 2-ой вершины')

# третяя вершина / нижняя точка рассматриваемого участка графика
plt.plot(vertex_3_x, vertex_3_y, 'rx')

# убывает от второй до третей вершины
neg_down_2 = np.arange(vertex_2_x, vertex_3_x, 0.01)
plt.plot(neg_down_2, func_1(neg_down_2), 'c-', label='↓ от 2-ой до 3-ей вершины')

# возрастает от третьей вершины до второго корня
neg_up_2 = np.arange(vertex_3_x, root_2, 0.01)
plt.plot(neg_up_2, func_1(neg_up_2), 'b-', label='↑ от 3-ей вершины до 2-го корня')

# возрастает при y > 0
pos_down_2 = np.arange(root_2, x_right, 0.01)
plt.plot(pos_down_2, func_1(pos_down_2), 'm-', label='↑ при y > 0')


plt.legend()
plt.grid()
```
